### Pivot Function

```python
import pandas as pd
import numpy as np
```

```python
stocks_df = pd.read_csv("stock.csv")
```

```python
stocks_df
```

```python
# pivot
# i am interested in the closing price for each stock across the trading days

stocks_pivot = stocks_df.pivot(index="Company", columns="Date", values="Close")
stocks_pivot
```

```python
# all the columns
stocks_df.pivot(index="Company", columns="Date")
```

### Melt
- The melt function is a function that can be used to unpivot a DataFrame from wide format to long format, optionally leaving identifier variables set. This function is useful to massage a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns, considered measured variables (value_vars), are “unpivoted” to the row axis, leaving just two non-identifier columns, ‘variable’ and ‘value’.

Screenshot 2023-08-30 183251.jpg

```python
company_df = pd.read_csv("reshape.csv")
company_df
```

```python
company_melt = pd.melt(company_df, id_vars=["Name"])
company_melt
```

```python
pd.melt(company_df, id_vars=["Name"], value_vars=["Dept"])
```

```python
# customize variable and value names
pd.melt(company_df, id_vars=["Name"], value_vars=["Dept"], var_name="What Variable?", value_name="Which Dept?")
```

### Stack & Unstack
- both is used on a multi indexed dataframe


Stack
- To reshape from wide to long format


Unstack
- to reshape from long to wide



Screenshot 2023-08-30 211554.jpg

```python
# creating a multi index dataframe

header = pd.MultiIndex.from_product([['Before Upskilling','After Upskilling'],['Age','Salary']])
d=([[25,45000,27,56000],[38,89000,41,102000],[30,67000,31,70000],[45,130000,48,160000]])


upskilling_df = pd.DataFrame(d,
                  index=['Marcus','Anthony','Mason','Daniel'],
                  columns=header)
upskilling_df
```

```python
# stacking - wide to long

upskill_stacked = upskilling_df.stack()
upskill_stacked
```

```python
# level 0 stacking
upskilling_df.stack(level=0)
```

```python
# unstacking - long to wide
upskill_stacked.unstack()
```

### Pivot_Table


*   Pivot table function aggregates along with reshaping.
*   Works only with numeric values, non-numeric values will be dropped
* Mean is the default aggregation function, can be changed to other function



```python
stocks_pivottable = pd.pivot_table(stocks_df, index="Company")
stocks_pivottable
```

```python
pd.pivot_table(stocks_df, index=["Company", "Date"])
```

```python
pd.pivot_table(stocks_df, index="Company", aggfunc=np.sum, values="Volume") # changing the aggregation function like sum, mean etc
```

```python
pd.pivot_table(stocks_df, index="Company", aggfunc=[np.amin, np.amax])
```
