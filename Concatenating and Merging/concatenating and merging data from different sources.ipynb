# Concatenating and Merging Data from Different Sources

```python
import pandas as pd
```

```python
# Concatenating Data Sets


df1_dummy= {
          'serial_id':['1','2','3','4','5'],
          'sale_month': ['Jan','Feb','Mar','Apr','May'],
          'sales': ['12300','25100','17800','20100','2100']
}

df1=pd.DataFrame(df1_dummy, columns=['sales', 'sale_month'])
df1
```

```python
df2_dummy= {
          'serial_id':['6','7','8','9','10'],
          'sale_month': ['Jun','Jul','Aug','Sep','Oct'],
          'sales': ['2500','23700','24600','2400','23950']
}

df2=pd.DataFrame(df2_dummy, columns=['sales', 'sale_month'])
df2
```

```python
df3_dummy= {
          'sales_threshold':['No','Yes','No','Yes','Yes','Yes','Yes','Yes','Yes','Yes'],
          'bonus_threshold':['No','Yes','No','No','No','Yes','No','Yes','Yes','No']
}
df3=pd.DataFrame(df3_dummy, columns=['sales_threshold','bonus_threshold'])
df3



```

```python
#concatenating the first and second df row-wise
df_row=pd.concat( [df1,df2]  , ignore_index=True ) # the ignore_index=True is for fixing the indexes numbering
df_row
```

```python
#concatenating the new DF with the 3rd df

df_full=pd.concat([df_row, df3], axis=1)
df_full
```

#Merging Datasets
- combines dataframes by aligning on columns.
- the process of joining two dataframes together based on some common columns
-in joining we use display instead of print
- in joining we enclose the dictionary with another ()

Types are inner, left, right outer



```python
#one to one join
#be sure to enclose the dictionary sets with another ()
df_1a= pd.DataFrame(
         {
          'product':['Prod_1','Prod_2','Prod_3','Prod_4'],
          'division':['Div_A','Div_B','Div_C','Div_D']
          }
    )


df_1b=pd.DataFrame({
    'sales': ['12300','25100','17800','20100'],
    'product':['Prod_3','Prod_2','Prod_4','Prod_1']
})

display(df_1a,df_1b)
```

```python
df_1ab=pd.merge(df_1a,df_1b)
df_1ab
```

```python

# Many to one joins

df_1c=pd.DataFrame (
    {
      'division':['Div_A','Div_B','Div_C'],
      'manager':['Roger','Rafael','Novak']
    }
)

df_1c
```

```python
df_1abc=pd.merge(df_1ab,df_1c)
df_1abc
```

```python
#many to many joins


df_1d = pd.DataFrame({"division" : ["Div_A", "Div_A", "Div_B", "Div_C", "Div_C", "Div_C"],
                   "emp_grade" : ["13", "14+", "12", "11", "10", "9-"]})

df_1d
```

note that the new data frame contains duplicates to the first dummy data frame which also contains duplicates. But with merge the data were cleaned

```python
df_1ad = pd.merge(df_1a, df_1d)

df_1ad

```

#Merge Keys

On parameter
*   Explicitly specific the name of the key column



Left_on, right_on parameters
* for merging two datasets with differnt columns names, but similar data


Left_index, right_index parameters
* for merging on an index rather than a column


```python
# on parameter
df_1a= pd.DataFrame(
         {
          'product':['Prod_1','Prod_2','Prod_3','Prod_4'],
          'division':['Div_A','Div_B','Div_C','Div_D']
          }
    )


df_1b=pd.DataFrame({
    'sales': ['12300','25100','17800','20100'],
    'product':['Prod_3','Prod_2','Prod_4','Prod_1']
})


display(pd.merge(df_1a, df_1b, on="product"))
```

```python
# left_on, right_on
df1_new = pd.DataFrame({"project" : ["Prod_1", "Prod_2", "Prod_3", "Prod_4"],
                   "division": ["Div_A", "Div_B", "Div_C", "Div_B"],
                   })

display(pd.merge(df1_new, df_1b, left_on="project", right_on="product"))
```

```python
#because of the redundancy in the output dataframe we have to drop one column.

display(pd.merge(df1_new, df_1b, left_on="project", right_on="product").drop("product", axis=1))
```

```python
# setting index
df1= pd.DataFrame(
         {
          'product':['Prod_1','Prod_2','Prod_3','Prod_4'],
          'division':['Div_A','Div_B','Div_C','Div_D']
          }
    )


df2=pd.DataFrame({
    'sales': ['12500','10800','5600','7900'],
    'product':['Prod_3','Prod_2','Prod_4','Prod_1']
})


df1_index = df1.set_index("product")
df2_index = df2.set_index("product")

display(df1_index, df2_index)
```

```python
display(pd.merge(df1_index, df2_index, left_index=True, right_index=True))
```

# the how Parameter


```python
# dummy dataframes with actor names from the moves Ironman
# and Avengers: End Game

df_a = pd.DataFrame({"id" : ["1", "2", "3", "4"],
                   "actor_first_name": ["Robert", "Gwyneth", "Jon", "Paul"],
                   "actor_last_name" : ["Downey Jr.", "Paltrow", "Favreau", "Bettany"],
                   "value" : ["10", "6", "7", "7"]
                   })

df_b = pd.DataFrame({"id" : ["1", "2", "3", "4", "5", "6"],
                    "actor_first_name" : ["Robert", "Chris", "Chris", "Mark", "Scarlett", "Jeremy"],
                    "actor_last_name" : ["Downey Jr.", "Evans", "Hemsworth", "Ruffalo", "Johansson", "Renner"]
                    })

display(df_a, df_b)
```

```python
# inner join
display(pd.merge(df_a, df_b, how="inner"))
```

```python
# left join
display(pd.merge(df_a, df_b, how="left"))
```

```python
# right join
display(pd.merge(df_a, df_b, how="right"))
```

```python
# outer join
display(pd.merge(df_a, df_b, how="outer", indicator=True))
```

```python
df_s1 = pd.DataFrame({"product" : ["Prod_1", "Prod_2", "Prod_3", "Prod_4"],
                   "division": ["Div_A", "Div_B", "Div_C", "Div_D"],
                   })

df_s2 = pd.DataFrame({"product" : ["Prod_1", "Prod_2", "Prod_3", "Prod_4"],
                   "division": ["Div_C", "Div_A", "Div_B", "Div_D"],
                   })

display(pd.merge(df_s1, df_s2, on="product"))
```

```python
# custom suffix

display(pd.merge(df_s1, df_s2, on="product", suffixes=["_LD", "_RD"]))
```
